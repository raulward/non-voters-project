### **3.1 Sobre o público da pesquisa** <h3>

**Qual o perfil da amostra analisada? Verifique a distribuição da idade, gênero, raça, nível educacional e salarial.**

**Distribuição de idade, raça, gênero e educação**

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 6))

sns.boxplot(data=df_non_voters, x="ppage", ax=axes[0, 0])
axes[0, 0].set_title("Distribuição de Idade")
axes[0, 0].set_xlabel("Idade")
axes[0, 0].set_ylabel("Contagem")

sns.countplot(data=df_non_voters, x="race", ax=axes[0, 1])
axes[0, 1].set_title("Distribuição de Raça")
axes[0, 1].set_xlabel("Raça")
axes[0, 1].set_ylabel("Contagem")

sns.countplot(data=df_non_voters, x="gender", ax=axes[1, 0])
axes[1, 0].set_title("Distribuição de Gênero")
axes[1, 0].set_xlabel("Gênero")
axes[1, 0].set_ylabel("Contagem")

sns.countplot(data=df_non_voters, x="educ", ax=axes[1, 1])
axes[1, 1].set_title("Distribuição de Educação")
axes[1, 1].set_xlabel("Educação")
axes[1, 1].set_ylabel("Contagem")

plt.tight_layout()
plt.show()
```

Ao analisar as distruibuições, notamos alguns pontos.

1. Faixa etária do grupo entrevistado está concentrada entre 35 e 65 anos aproximadamente;
2. Dos eleitores que responderam a pesquisa, a maior parte é da raça branca;
3. Existe uma concentração semelhante de homens e mulheres no grupo entrevistado;
4. A maior parte dos entrevistados não possui uma graduação completa.

**Distribuição de renda**

```python
labels = ["Less than $40k", "$40-75k", "$75-125k", "$125k or more"]
sns.countplot(data=df_non_voters, x="income_cat", order=labels)
plt.title("Distribuicao de Renda")
plt.xlabel("Renda")
plt.ylabel("Contagem")
plt.show()
```

```python
df_non_voters_abs_freq = df_non_voters["income_cat"].value_counts()
df_non_voters_abs_rel = df_non_voters["income_cat"].value_counts(normalize=True) * 100

df_tab_freq_rel = pd.DataFrame({"Tabela de frequência absoluta": df_non_voters_abs_freq, "Tabela de frequência relativa": df_non_voters_abs_rel})

df_tab_freq_rel
```

A renda se encontra bem distribuída, com a maior parte dos entrevistados possuindo uma entre 75-125 mil doláres

**Assumindo que a amostra representa adequadamente a população americana, qual a raça que possui maior poder aquisitivo? E menor?**

```python
cross_tab_income_x_race = pd.crosstab(df_non_voters["income_cat"], df_non_voters["race"], normalize="index") * 100

cross_tab_income_x_race.reset_index(inplace=True)

df_cross_tab = cross_tab_income_x_race.melt(id_vars="income_cat", var_name="race", value_name="percent")

plt.figure(figsize=(10,6))

labels = ["Less than $40k", "$40-75k", "$75-125k", "$125k or more"]
sns.barplot(data=df_cross_tab, x="income_cat", y="percent", hue="race", palette="viridis", order=labels)

plt.title("Distribuição Percentual do Poder Aquisitivo por Raça")
plt.xlabel("Faixa de Renda")
plt.ylabel("Percentual (%)")
plt.xticks(rotation=45)
plt.legend(title="Raça")
```

Aqui nota-se uma disparidade entre as raças no quesito renda. Com a maioria da população negra e hispânica estando incluída em grupos de salários menores. Além disso, boa parte da população branca possui rendas maiores, representando a maioria dos entrevistados que possuem salários acima de 125k. Entretanto, vale ressaltar que essa disparidade também se dá pela diferença na quantidade entre entrevistados brancos e de outras raças.

**Ainda assumindo que temos uma amostra representativa, podemos dizer que nível educacional e poder aquisitivo estão correlacionados?**

```python
plt.figure(figsize=(10, 6))

cross_tab_income_x_educ = pd.crosstab(df_non_voters["income_cat"], df_non_voters["educ"], normalize="index") * 100

cross_tab_income_x_educ.reset_index(inplace=True)

df_income_x_educ = cross_tab_income_x_educ.melt(id_vars="income_cat", var_name="educ", value_name="percent")


labels = ["Less than $40k", "$40-75k", "$75-125k", "$125k or more"]
sns.barplot(df_income_x_educ, x="income_cat", y="percent", hue="educ", palette="viridis", order=labels)
plt.title("Distribuiçao percentual Poder aquisitivo x Educação")
plt.xlabel("Faixa Salarial")
plt.ylabel("Percentual (%)")
plt.xticks(rotation=45)
plt.legend(title="Nível educacional")
```

Nota-se que níveis educacionais mais altos indicam salários mais altos, com cerca de 68% dos entrevistados presentes na maior faixa salarial possuindo ao menos uma graduação.

### **3.2 O que é ser considerado um bom americano?** <h3>

**Um nível educacional maior implica em maior responsabilidade em relação a votar?**

```python
plt.figure(figsize=(10, 6))

cross_tab_educ_x_voterc = pd.crosstab(df_non_voters["educ"], df_non_voters["voter_category"], normalize="index") * 100

cross_tab_educ_x_voterc.reset_index(inplace=True)

df_educ_x_voter_category = cross_tab_educ_x_voterc.melt(id_vars="educ", var_name="voter_category", value_name="percent")

labels = ["High school or less", "Some college", "College"]
sns.barplot(df_educ_x_voter_category, x="educ", y="percent", hue="voter_category", palette="viridis", order=labels)
plt.title("Distribuiçao percentual Educação x Frequência de votos")
plt.xlabel("Nível educacional")
plt.ylabel("Percentual (%)")
plt.xticks(rotation=45)
plt.legend(title="Frequência de voto")
```

Nota-se que a escolaridade não é um índice forte de regularidade de voto, logo a maioria dos entrevistados votam esporadicamente.

```python
filtered_mask = df_non_voters["race"] != "White"

df_filtered = df_non_voters[filtered_mask]
df_filtered = df_filtered[["Q2_1", "Q2_4", "Q2_7"]]

df_melt = df_filtered.melt(var_name="subquestion", value_name="importance_level")

count_data = df_melt.groupby(["subquestion", "importance_level"]).size().reset_index(name="count")

count_data = count_data[count_data["importance_level"] != -1]

count_data["percent"] = count_data.groupby("subquestion")["count"].transform(lambda x: x / x.sum() * 100)
```

```python
count_data["importance_level"].unique()
```

```python
def importance_lev_group(importance_level):
    if importance_level == 1:
        return "Very Important"
    elif importance_level == 2:
        return "Somewhere important"
    elif importance_level == 3:
        return "Not so important"
    elif importance_level == 4:
        return "Not at all important"


count_data["importance_level_group"] = count_data["importance_level"].apply(importance_lev_group)
```

**Dentre os não-brancos, o que é mais importante para ser um bom americano: votar em eleições, demonstrar a bandeira americana ou apoiar o exército? Existe uma diferença na preferência de não-brancos de maior poder aquisitivo vs menor poder aquisitivo?**

```python
plt.figure(figsize=(20, 6))

legend_label = ["Very Important", "Somewhere important", "Somewhere important", "Not at all important"]

sns.barplot(data=count_data, x="subquestion", y="percent", hue="importance_level_group", palette="viridis")
plt.title("Distribuição percentual das respostas sobre ser um bom cidadão")
plt.xticks(ticks=range(len(count_data["subquestion"].unique())), labels=["Voting in Elections", "Displaying American Flag", "Supporting the Military"])
plt.legend(title="Nível de importancia")
plt.xlabel("Perguntas")
plt.ylabel("Percentual de resposta")
plt.show()
```

Votar nas eleições é considerado dentre os entrevistados não-brancos um indício de ser um bom cidadão. Além disso, quando questionados a respeito do suporte ao exército, quase 60% dos entrevistados considera muito importante

```python
def categorize_age(age):
    if age >= 18 and age <= 30:
        return "18-30"
    elif age >= 31 and age <= 50:
        return "31-50"
    else:
        return "51+"

df_non_voters["age_group"] = df_non_voters["ppage"].apply(categorize_age)

df_tolerance = df_non_voters[["age_group", "Q2_8"]]

count_tolerance = df_tolerance.groupby(["age_group", "Q2_8"]).size().reset_index(name="count")

count_tolerance = count_tolerance[count_tolerance["Q2_8"] != -1]

count_tolerance["percent"] = count_tolerance.groupby("age_group")["count"].transform(lambda x: x/x.sum() * 100)

count_tolerance
```

Considerando que respeitar a opinião dos outros (Q2_8) é um indício de tolerância, ao compararmos as gerações de 18-30 anos, 31-50 anos e +51, qual é a mais tolerante e a menos tolerante?

```python
plt.figure(figsize=(10, 6))

count_tolerance["questions_group"] = count_tolerance["Q2_8"].apply(importance_lev_group)

sns.barplot(data=count_tolerance, x="age_group", y="percent", hue="questions_group", palette="viridis")
plt.title("Importância da tolerância x Grupo de idade")
plt.xlabel("Grupo de idade")
plt.ylabel("Percentual (%)")
plt.legend(title="Nível de importância")
plt.show()
```

Ao analisar os entrevistados a respeito da importância da tolerância, observamos que a faixa etária mais tolerante (considerando que a pergunta é um fator que determina a tolerância) é a de 51+, seguida pelos entrevistados que possuem entre 31-50 anos e os eleitores mais jovens, categorizados entre 18-30 anos.

### **3.3 Escolhas partidárias** <h3>

**Qual escolha partidária, incluindo pessoas sem partido (Q30), possui eleitores mais jovens? E qual possui mais mulheres como apoiadoras?**

```python
df_voter_choice = df_non_voters[["age_group", "Q30"]]

df_voter_choice = df_voter_choice[df_voter_choice["age_group"] == "18-30"] 

df_voter_choice = df_voter_choice.groupby(["age_group", "Q30"]).size().reset_index(name="count")

df_voter_choice = df_voter_choice[df_voter_choice["Q30"] != -1]

df_voter_choice["percent_partidary_choice_age"] = df_voter_choice.groupby("age_group")["count"].transform(lambda x: x/x.sum() * 100)

df_voter_choice
```

```python
def partidary_choice(num):
    if num == 1:
        return "Republican"
    elif num == 2:
        return "Democrat"
    elif num == 3:
        return "Independent"
    elif num == 4:
        return "Another partidary choice"
    else:
        return "No preference"
```

```python
plt.figure(figsize=(20, 6))


df_voter_choice["Q30"] = df_voter_choice["Q30"].apply(partidary_choice)

sns.barplot(data=df_voter_choice, x="Q30", y="percent_partidary_choice_age", dodge=True, errorbar=None)

plt.title("Escolha Partidária por Grupo de Idade")
plt.xlabel("Grupo de Idade")
plt.ylabel("Percentual (%)")
plt.legend(title="Partido Político")
plt.show()
```

O partido que possui mais eleitores jovens, descritos na faixa etária entre 18-30 anos é o partido democrata.

```python
df_voter_choice_g = df_non_voters[["gender", "Q30"]]

df_voter_choice_g = df_voter_choice_g.groupby(["gender", "Q30"]).size().reset_index(name="count")

df_voter_choice_g = df_voter_choice_g[df_voter_choice_g["Q30"] != -1]

df_voter_choice_g["percent_partidary_choice_gender"] = df_voter_choice_g.groupby("gender")["count"].transform(lambda x: x/x.sum() * 100)

df_voter_choice_g["Q30"] = df_voter_choice_g["Q30"].apply(partidary_choice)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df_voter_choice_g, x="gender", y="percent_partidary_choice_gender", hue="Q30", palette="viridis")

plt.title("Escolha Partidária por gênero")
plt.xlabel("Grupo de Idade")
plt.ylabel("Percentual (%)")
plt.legend(title="Partido Político")
plt.show()
```

Nota-se que o partido que contem mais eleitores tanto homens, quanto mulheres é o partido democrata
